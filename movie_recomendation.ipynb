```
import ast
#function to take all name from data
def converter(df1):
  list1=[]
  for i in ast.literal_eval(df1):
    list1.append(i['name'])
  return list1
```

```
df['genres']=df['genres'].apply(converter)
```

```
#converting keywords column
df['keywords']=df['keywords'].apply(converter)
```

```
import ast
#function to take top 3 name from data
def converter1(df1):
  list1=[]
  k=0
  for i in ast.literal_eval(df1):
    if k !=3:
      list1.append(i['name'])
      k+=1
    else:
      break
  return list1
df['cast']=df['cast'].apply(converter1)
```

```
import ast
#function to take top 3 name from data
def converter2(df1):
  list1=[]
  for i in ast.literal_eval(df1):
    if i['job'] == 'Director':
      list1.append(i['name'])

  return list1
df['crew']=df['crew'].apply(converter2)
```

```
df.head(2)
```

```
#converting overview string column into list
df['overview']=df['overview'].apply( lambda x:x.split())
```

```
df.sample(2)
```

```
#we are replaccing space from word and converting into single words so our recomendation system don't confuse
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```
df.head(3)
```

```
df.columns
```

```
#concatinate overview', 'genres', 'keywords', 'cast', 'crew'
df['tag']=df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```
df.head()
```

```
new_df=df[['movie_id','title','tag']]
```

```
new_df.head(2)
```

```
#converting tag list into string
new_df['tag']=new_df['tag'].apply(lambda x:" ".join(x))
```

```
new_df.head(2)
```

```
#converting data.tag into lower case
new_df['tag']=new_df['tag'].apply(lambda x:x.lower())
new_df.head(2)
```

```
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```
def stem(text):
  y=[]
  for i in text.split():
      y.append(ps.stem(i))
  return " ".join(y)
```

```
new_df['tag']=new_df['tag'].apply(stem)
```

```
new_df['tag'][0]
```

```
#bag of words is used to combine data into string and from string we check the count of most occurance word and after that we take 1st row and check same words , give count as 2,3,4  repeted word so we can plot the vector on graph and check the nearest vectors

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```
vector=cv.fit_transform(new_df['tag']).toarray()
```

```
#cosine library is used to check the distance between two raw data like 10 degree is minimum distance or 90/100 degree is high distance
from sklearn.metrics.pairwise import cosine_similarity
```

```
similarity=cosine_similarity(vector)
```

```
similarity
```

```
# main function to recomend top 5 similar content movie
def recomended(movie):
  # we are takig title of data and if title == moive(means name of movie which is given by user)is equal then we get the index value of movie ex. avatar is 1st movie in title column then index value will be 0 as output
  movie_index=new_df[new_df['title']== movie].index[0]
  #similarity is the distance matrix of every movie tag
  distance=similarity[movie_index]
  #enumerate is used to add index and reverse is used to sort list from high to low after that using lambda to take positional value as 1 ,we hae 0,1 in list as title no. in list and distance ,so we are taking top 5 movie
  movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    # taking index value as i[0] and print title name of movie
    print(new_df.iloc[i[0]].title)
```

```
recomended("Pirates of the Caribbean: At World's End")
```
